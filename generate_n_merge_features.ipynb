```python
import numpy as np
import pandas as pd
```

```python
weather_df = pd.read_csv('archive odesa.csv', delimiter=',')
weather_df.head(10)
```

```python
weather_df.columns
```

```python
features = ['time', 'temperature_2m (°C)', 'relativehumidity_2m (%)', 'apparent_temperature (°C)', 
            'snowfall (cm)', 'dewpoint_2m (°C)', 'cloudcover (%)', 'windspeed_10m (km/h)', 'windgusts_10m (km/h)']
weather_df = weather_df[features]
weather_df.head()
```

```python
test_df = pd.read_csv('final_test_with_sequencesv2.csv', delimiter=',')
test_df.head()
```

```python
orders_df = pd.read_csv('orders_time_features.csv', delimiter=',')
orders_df.drop(columns=orders_df.columns[0], inplace=True)
orders_df.head()
```

```python
data = []
weather_datetime = pd.DatetimeIndex(weather_df['time'])
orders_datetime = pd.DatetimeIndex(test_df['running_time'])
for order_id, dt in zip(test_df['Id'], orders_datetime):
    idx = (weather_datetime - dt).days.searchsorted(0, side='left')
    append_data = [order_id] + weather_df.iloc[idx-1, 1:].to_list()
    data.append(append_data)
```

```python
join_df = pd.DataFrame(data, columns=['Id'] + features[1:])
join_df.head()
```

```python
final_df = pd.merge(test_df, join_df, on='Id', how='inner')
final_df.head()
```

```python
final_df.to_csv('final_test_with_sequencesv2.csv', index=False)
```
