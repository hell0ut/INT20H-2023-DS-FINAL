```python
start_node_lat = df1['sequence'].map(lambda x: df3[df3['node_id'] == int(x.strip('][').split(', ')[0])]['lat'].values[0])
start_node_lon = df1['sequence'].map(lambda x: df3[df3['node_id'] == int(x.strip('][').split(', ')[0])]['lon'].values[0])
end_node_lat = df1['sequence'].map(lambda x: df3[df3['node_id'] == int(x.strip('][').split(', ')[-1])]['lat'].values[0])
end_node_lon = df1['sequence'].map(lambda x: df3[df3['node_id'] == int(x.strip('][').split(', ')[-1])]['lon'].values[0])
```

```python
df1['start_node_lat'] = start_node_lat
df1['start_node_lon'] = start_node_lon
df1['end_node_lat'] = end_node_lat
df1['end_node_lon'] = end_node_lon
```

```python
AVG_EARTH_RADIUS = 6371  # in km
def haversine_array(lat1, lon1, lat2, lon2):
    lat1, lon1, lat2, lon2 = map(np.radians, (lat1, lon1, lat2, lon2))
    lat = lat2 - lat1
    lon = lon2 - lon1
    d = np.sin(lat * 0.5) ** 2 + np.cos(lat1) * np.cos(lat2) * np.sin(lon * 0.5) ** 2
    h = 2 * AVG_EARTH_RADIUS * np.arcsin(np.sqrt(d))
    return h

def dummy_manhattan_distance(lat1, lon1, lat2, lon2):
    a = haversine_array(lat1, lon1, lat1, lon2)
    b = haversine_array(lat1, lon1, lat2, lon1)
    return a + b

def bearing_array(lat1, lon1, lat2, lon2):
    lng_delta_rad = np.radians(lon2 - lon1)
    lat1, lon1, lat2, lon2 = map(np.radians, (lat1, lon1, lat2, lon2))
    y = np.sin(lng_delta_rad) * np.cos(lat2)
    x = np.cos(lat1) * np.sin(lat2) - np.sin(lat1) * np.cos(lat2) * np.cos(lng_delta_rad)
    return np.degrees(np.arctan2(y, x))
```

```python
df1.loc[:, 'distance_haversine'] = haversine_array(df1['start_node_lat'].values, df1['start_node_lon'].values, 
                                                  df1['end_node_lat'].values, df1['end_node_lon'].values)
df1.loc[:, 'distance_dummy_manhattan'] = dummy_manhattan_distance(df1['start_node_lat'].values, df1['start_node_lon'].values, 
                                                  df1['end_node_lat'].values, df1['end_node_lon'].values)
df1.loc[:, 'direction'] = bearing_array(df1['start_node_lat'].values, df1['start_node_lon'].values, 
                                                  df1['end_node_lat'].values, df1['end_node_lon'].values)
```

```python
from sklearn.decomposition import PCA
coords = np.vstack((df2[['start_node_lat', 'start_node_lon']].values,
                    df2[['end_node_lat', 'end_node_lon']].values))

pca = PCA().fit(coords)
df2['pickup_pca0'] = pca.transform(df2[['start_node_lat', 'start_node_lon']])[:, 0]
df2['pickup_pca1'] = pca.transform(df2[['start_node_lat', 'start_node_lon']])[:, 1]
df2['dropoff_pca0'] = pca.transform(df2[['end_node_lat', 'end_node_lon']])[:, 0]
df2['dropoff_pca1'] = pca.transform(df2[['end_node_lat', 'end_node_lon']])[:, 1]
```

```python
from sklearn.cluster import MiniBatchKMeans

sample_ind = np.random.permutation(len(coords))[:500000]
kmeans = MiniBatchKMeans(n_clusters=100, batch_size=10000).fit(coords[sample_ind])

df2.loc[:, 'start_cluster'] = kmeans.predict(df2[['start_node_lat', 'start_node_lon']])
df2.loc[:, 'end_cluster'] = kmeans.predict(df2[['end_node_lat', 'end_node_lon']])
```

```python
start_cluster_ohe = pd.get_dummies(df2.start_cluster, prefix='s', prefix_sep='_')
end_cluster_ohe = pd.get_dummies(df2.end_cluster, prefix='e', prefix_sep='_')
```

```python
df2.drop(columns=['Id', 'running_time', 'completed_time', 'start_cluster', 'end_cluster'], inplace=True)
```

```python
df2 = df2.join(start_cluster_ohe).join(end_cluster_ohe)
df2.head()
```
